#### Bar chart with individual bar colors

```python
fig , ax = plt.subplots()

fruits =['apple' , 'strawberry' , 'banana' , 'mango' ]
counts = [40, 100, 75, 90]
bar_label = ['red' , 'pink' , 'yellow' , 'orange']
bar_colors = ["#F01010" , "#E60C79" , "#F0F010" , "#FDAB13"]
ax.bar(fruits , counts , color = bar_colors)
ax.bar(fruits , counts , label = bar_label , color = bar_colors)
ax.set_ylabel("Fruit counts")
ax.set_title("Fruit count analysis")
plt.legend(title = "Fruit Colors"  )
plt.show()
```

### Pie Chart

 It is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents.

```python
import matplotlib.pyplot as plt

plt.pie(x = [25,67,12,45,77] , labels = ['malaika' , 'alice' , 'emma' , 'sofi' , 'Bob'] , colors=['red' , 'blue' , 'green' , 'yellow' , 'orange'])
plt.title("Student Marks Distribution" , size = 21)

plt.show()
```

##### with auto-label slices
##### automatic percentage text on pie chart slices.It formats and displays percent values for each slice
##### Meaning of %1.1f%%:
##### f → float number
##### .1 → 1 digit after decimal
##### %% → shows % symbol

```python
import matplotlib.pyplot as plt

plt.pie(x = [25,67,12,45,77] , labels = ['malaika' , 'alice' , 'emma' , 'sofi' , 'Bob'] , colors=["#292EC5" , "#B92929" , "#1F430F" , "#6D2DCE" , "#E06703"] , autopct='%1.1f%%')
plt.title("Student Marks Distribution" , size = 21)

plt.show()
```

### Hatch slice
##### It is used to fill the pie chart slices with different patterns to distinguish between different categories.

```python
import matplotlib.pyplot as plt

plt.pie(x = [25,67,12,45,77] , labels = ['malaika' , 'alice' , 'emma' , 'sofi' , 'Bob'] , colors=["#292EC5" , "#B92929" , "#1F430F" , "#6D2DCE" , "#E06703"] , hatch = ['/' , 'x', 'O.O', '*' ,'.||.'])
plt.title("Student Marks Distribution" , size = 21)

plt.show()
```

#### Swap the autopact and labels

```python
import matplotlib.pyplot as plt

labels = ['malaika' , 'alice' , 'emma' , 'sofi' , 'Bob']
sizes = [25,67,12,45,77]
fig , ax = plt.subplots()
ax.pie(sizes , labels= labels , colors=["#292EC5" , "#B92929" , "#1F430F" , "#6D2DCE" , "#E06703"] , autopct='%1.1f%%' , labeldistance=0.50,pctdistance=1.25  )
```

##### Explode , shade and rotate slices


```python
import matplotlib.pyplot as plt

plt.pie(x = [25,67,12,45,77] , labels = ['malaika' , 'alice' , 'emma' , 'sofi' , 'Bob'] , colors=["#292EC5" , "#B92929" , "#1F430F" , "#6D2DCE" , "#E06703"] , autopct='%1.1f%%' , explode = (0.0,0.0 , 0.2 , 0.0, 0.1) , shadow = True , startangle=90)
plt.title("Student Marks Distribution" , size = 18)


plt.show()
```

### Histograms

 A histogram is a visual way to show how data is distributed. It displays the frequency (how often) values appear in your data by grouping them into bins (ranges) and drawing bars.Key Parameters

bins – Number of bins/ranges (more bins = more detail)
color – Bar color
edgecolor – Border color of bars
alpha – Transparency (0-1)

```python
import matplotlib.pyplot as plt

# Sample data: test scores
scores = [45, 67, 89, 23, 56, 78, 90, 34, 67, 88, 92, 55, 72, 88, 91]

# Create histogram
plt.hist(scores, bins=5)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Test Score Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Normal distribution = perfect tower shape
data = np.random.normal(loc=70, scale=10, size=1000)

plt.hist(data, bins=30, color='green', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Perfect Tower')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=70, scale=10, size=1000)

plt.hist(data, bins=30, color=np.random.rand(30, 3), edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Multicolor Tower')
plt.show()
```
